# Desafio do Kaggle sobre o pre√ßo de casa.

```python
# Importando os pacotes

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
%matplotlib inline
pd.set_option("display.max_columns", 200)
```

```python
# Lendo os arquivos

sample = pd.read_csv('datasets/sample_submission.csv')
train = pd.read_csv('datasets/train.csv')
test = pd.read_csv('datasets/test.csv')
```

```python
preco = train.iloc[:,80]
# df = train.copy()
df = train.iloc[:,:80]
```

```python
corr_mat = df.corr()
corr_mat['SalePrice'].sort_values(ascending=False)
```

# Tratando os dados

```python
df.isnull().sum()
df.fillna(0, inplace=True)
```

```python
def tratamento(data):
    for x in data:
        if data[x].dtype == np.object:
            cov, ind = data[x].factorize()
            data[x] = cov
```

```python
tratamento(df)
```

```python
df.head()
```

# Preparando a base de dados

```python
# Separando os dados ja tratados
df_trat = df.iloc[:,:12]
```

```python
df_trat.isnull().sum()
```

# Preparando a base de teste

```python
df_test = test.copy()
```

```python
df_test.fillna(0, inplace=True)
```

```python
tratamento(df_test)
```

```python
df_test
```

# Preparando o modelo

```python
lin_reg = LinearRegression()
lin_reg.fit(df, preco)
```

```python
reg_trat = lin_reg.predict(df_test)
```

```python
test['SalePrice'] = reg_trat
sample_test = pd.DataFrame(test['SalePrice'])
sample_test.index = test['Id']
sample_test.to_csv('sample_submission6.1.csv')
```

# Gradiente Descendente

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(n_iter=50, penalty=None, eta0=0.1)
sgd_reg.fit(df_trat, preco.ravel())
```

```python
sgd_pred = sgd_reg.predict(df_test)
```

```python
test['SalePrice'] = sgd_pred
sample_test = pd.DataFrame(test['SalePrice'])
sample_test.index = test['Id']
sample_test.to_csv('sample_submission6.csv')
```
