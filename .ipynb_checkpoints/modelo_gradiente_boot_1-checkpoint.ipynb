```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('datasets/train.csv', index_col='Id')
test = pd.read_csv('datasets/test.csv', index_col='Id')
```

```python
x, y = train.drop('SalePrice', axis=1), train.SalePrice.copy()
```

```python
x1 = x.select_dtypes(include=[np.number]).fillna(-1)
x_test = test.select_dtypes(include=[np.number]).fillna(-1)
```

```python
x2 = train.copy()

for col in x2.columns:
    if x2[col].dtype == object:
        enc = LabelEncoder()
        x2[col] = enc.fit_transform(x[col].fillna('Missing'))
```

```python
df_test = test.copy()

for col in df_test.columns:
    if df_test[col].dtype == object:
        enc = LabelEncoder()
        df_test[col] = enc.fit_transform(test[col].fillna('Missing'))

df_test.fillna(-1, inplace=True)
```

```python
df = pd.concat([x1,x2])
df.fillna(0, inplace=True)
```

```python
x, y = df.drop('SalePrice', axis=1), df.SalePrice.copy()
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
model = GradientBoostingRegressor(random_state=0)
  
model.fit(x_train, y_train)
ptest = model.predict(df_test)
```

```python
prev = pd.Series(ptest, index= test.index, name='SalePrice')
prev.to_csv("modelo_gradiente_boot_3.csv", header=True)    
```
