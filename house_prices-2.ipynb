```python
# Importando os pacotes

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
%matplotlib inline
pd.set_option("display.max_columns", 200)
```

```python
# Lendo os arquivos

sample = pd.read_csv('datasets/sample_submission.csv')
train = pd.read_csv('datasets/train.csv')
test = pd.read_csv('datasets/test.csv')
```

```python
preco = train.iloc[:,80]
# df = train.copy()
df = train.iloc[:,:80]
```

```python
# corr_mat = df.corr()
# corr_mat['SalePrice'].sort_values(ascending=False)
```

# Tratando os dados

```python
# MSZoning: Identifies the general zoning classification of the sale.
class_zoneam = {'A':0, 'C':1, 'FV':2, 'I':3, 'RH':4,'RL':5,'RP':6,'RM':7}
df['MSZoning'] = df['MSZoning'].map(class_zoneam)
df['MSZoning'] = df['MSZoning'].fillna(8)

# steet Grvl,Pave
acesso = {'Grvl':0,'Pave':1}
df['Street'] = df['Street'].map(acesso)

# Tipo de acesso da propriedade
tipoAcesso = {'Grvl':0, 'Pave':1}
df['Alley'] = df['Alley'].map(tipoAcesso)
df['Alley'] = df['Alley'].fillna(3)

df['LotFrontage'] = df['LotFrontage'].fillna(0)

# Forma da propriedade
forma = {'Reg':0,'IR1':1,'IR2':2,'IR3':3}
df['LotShape'] = df['LotShape'].map(forma)

# planicidade
planicidade = {'Lvl':0,'Bnk':1,'HLS':2, 'Low':3}
df['LandContour'] = df['LandContour'].map(planicidade)

# tipos de utilitarios disponivel
utilit = {'AllPub':0, 'NoSewr':1, 'NoSeWa':2,'ELO':3}
df['Utilities'] = df['Utilities'].map(utilit)
```
